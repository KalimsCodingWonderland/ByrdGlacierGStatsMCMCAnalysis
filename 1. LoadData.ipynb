### Lets determine the high-velocity region, where we want to apply MCMC algorithm

The *highvel_mask* returned try to smoothly enclose a region of high velocity and excluding locations where the ice is not grounded. This region could be used later to constrain the sampling location of the MCMC.

The function works as first find high velocity region, then smooth the boundary of this region. Because this smoothing generally will shrink the boundary, the boundary is later expanded outward for *distance_max* meters.

The degree of smoothness of the boundary could be modified by the optional argument *smooth_mode*, which is default to 10. A higher smooth_mode will give a smoother boundary

The *ocean_mask* will be 1 when the location is ocean (open ocean water without ice, sea ice, or ice shelf), and will be 0 otherwise.

The *grounded_ice_mask* is 1 at location where ice is present and grounded, and will be 0 otherwise.

An example where *ocean_mask == 0 and grounded_ice_mask == 0* will be ice-free terrestrial land

__Q3: Change distance_max and smooth_mode parameters, how do you see the high velocity region changed by these parameter? Which parameter give a reasonable answer?__

Answer: 

Effect of smooth_mode. Increasing smooth_mode makes the high-velocity regionâ€™s boundary smoother and more rounded. Small holes or jagged peninsulas get removed/merged. If the value is too small, it leaves noisy and pixelated edges. If the value is too large, it over smooths and erases useful detail.

Effect of distance_max. This parameter sets how far the region expands outward. Larger values thicken the mask, and if the value is too large it makes the â€œhigh-velocity zoneâ€ look bigger than it really is.  

I think a distance_max of 7 and a smooth_mode of 20 gave the best output as it resemlbles the actual shape of the high velocity region the most (without over smoothing and without overestimating the region).

Notice, we aims to cover region with 50 m/yr velocity magnitude

```python
# 1) Build base masks
ocean_mask = (bedmap_mask == 3) | np.isnan(bedmap_mask)         # open ocean or floating
grounded_ice_mask = (bedmap_mask == 1) | (bedmap_mask == 2)     # grounded rock/ice
velocity_threshold = 50                                         # m/yr

# 2) Define an initial high-velocity core (on grounded ice)
velmag = np.hypot(velx, vely)
mask = grounded_ice_mask & (velmag >= velocity_threshold)

# 3) Include ocean/floating to â€œpunch holesâ€ (so borders get clean after smoothing)
mask = mask | ocean_mask

# 4) Smooth the edge so the border is clean and less jagged
smooth_mode = 20  # typical 10â€“40; bigger = smoother, slower
image = Image.fromarray((mask.astype(np.uint8) * 255))
image = image.filter(ImageFilter.ModeFilter(size=smooth_mode))
mask_mat = (np.array(image) // 255).astype(np.uint8)  # back to {0,1}

# 5) Expand outward a bit to cover trunk margins, then clip back to grounded ice
distance_max = 7  # number of grid cells (pixels), not meters
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (distance_max, 15))
dilate = cv2.dilate(mask_mat, kernel, iterations=1)

# 6) Final mask: dilated but restricted to grounded ice
highvel_mask = (dilate.astype(bool) & grounded_ice_mask)
```

```python
plt.pcolormesh(xx,yy,highvel_mask)
plt.axis('scaled')
```
